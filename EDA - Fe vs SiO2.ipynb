### Parece haber una disminución en el %Si en el concentrado a partir del 20/05/2017, ya que se observa una mayor densidad de los datos, posterior a dicha fecha, con valores principalmente por debajo del 3% de silicio en el concentrado.

```python
dict_rename = {
       '% Iron Feed':'%Fe Alim.', '% Silica Feed':'%Si Alim.', 'Starch Flow':'Almidon', 'Amina Flow':'Amina',
       'Ore Pulp Flow':'Flujo Pulpa', 'Ore Pulp pH':'pH Pulpa', 'Ore Pulp Density':'Densidad Pulpa',
       '% Iron Concentrate':'%Fe Conc.', '% Silica Concentrate':'%Si Conc.',
       'Flotation Column 01 Air Flow':'Air Flow 1', 'Flotation Column 02 Air Flow':'Air Flow 2',
       'Flotation Column 03 Air Flow':'Air Flow 3', 'Flotation Column 04 Air Flow':'Air Flow 4',
       'Flotation Column 05 Air Flow':'Air Flow 5', 'Flotation Column 06 Air Flow':'Air Flow 6',
       'Flotation Column 07 Air Flow':'Air Flow 7', 'Flotation Column 01 Level':'Cell Level 1',
       'Flotation Column 02 Level':'Cell Level 2', 'Flotation Column 03 Level':'Cell Level 3',
       'Flotation Column 04 Level':'Cell Level 4', 'Flotation Column 05 Level':'Cell Level 5',
       'Flotation Column 06 Level':'Cell Level 6', 'Flotation Column 07 Level':'Cell Level 7'
}

df = df.rename(columns=dict_rename)
df.head()
```

```python
# ANÁLISIS UNIVARIABLE: %Si Conc.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

df['%Si Conc.'].describe()
```

```python
# Separamos la variable dependiente de las independientes.

X = df.drop(columns=['%Si Conc.']) #Variables Independientes

Y = pd.DataFrame(df['%Si Conc.']) # Variable Dependiente
Y.tail()
```

```python
# Visualización de la distribución del %Si en el concentrado.
sns.histplot(df['%Si Conc.'], kde=True, stat='density')

#Realizamos la gráfica de una distribución normal para el %Si en el concentrado.
mu, std = stats.norm.fit(df['%Si Conc.'])
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)

skw = df['%Si Conc.'].skew()
krt = df['%Si Conc.'].kurt()

plt.text(3, 0.7, f'Skewness = {round(skw,2)}\nKurtosis = {round(krt,2)}', fontsize=15)

plt.show()
```

```python
# ANÁLISIS BIVARIABLE

# Diagrama de dispersión.
def scatterplot_func (x, y, **kwargs):
    sns.scatterplot(x=x, y=y, alpha=0.5)
    x = plt.xticks(rotation=90)

# Aplicamos un pivot al dataframe por cada variable independiente.
f = pd.melt(frame=df, id_vars=['%Si Conc.'], value_vars=X)

# Creamos las parcelas donde colocaremos las gráficas.
g = sns.FacetGrid(data=f, col='variable', col_wrap=2, sharex=False, sharey=False)

# Poblamos las parcelas con las gráficas.
g = g.map(scatterplot_func, 'value', '%Si Conc.')

plt.show()
```
